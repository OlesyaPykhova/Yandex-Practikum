## Модель прогнозирования оттока клиентов

### Разбиваем данные на обучающую и валидационную выборки

```python
# Разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop('churn', axis = 1)
y = df['churn']
```

```python
# Разбиваем данные на обучающую и валидационную выборки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

### Обучаем модель на train-выборке логистической регрессией


```python
from sklearn.linear_model import LogisticRegression

# Задаем алгоритм для нашей модели
lr_model = LogisticRegression(random_state=0)

# Обучаем модель
lr_model.fit(X_train, y_train)

# Делаем прогноз для объектов из валидационного набора данных
lr_predictions = lr_model.predict(X_test)
lr_probabilities = lr_model.predict_proba(X_test)[:,1]

warnings.filterwarnings('ignore')
```

### Обучаем модель на train-выборке случайным лесом

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Задаем алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)

# Обучаем модель случайного леса
rf_model.fit(X_train, y_train)

# Делаем прогнозы
rf_predictions = rf_model.predict(X_test)
rf_probabilities = rf_model.predict_proba(X_test)[:,1]
```

### Оценка метрик accuracy, precision и recall для обеих моделей на валидационной выборке

Оцените метрики accuracy, precision и recall для обеих моделей на валидационной выборке. Сравните по ним модели. Какая модель показала себя лучше на основании метрик? Не забудьте указать параметр random_state при разделении выборки и задании алгоритма.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score

# Функция,, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, y_proba, title = 'Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

```python
print_all_metrics(y_test, lr_predictions, lr_probabilities, title = 'Метрики для модели логистической регрессии:')
```

```python
print_all_metrics(y_test, rf_predictions, rf_probabilities, title = 'Метрики для модели случайного леса:')
```

Метрики для модели случайного леса показали себя лучше в сравнении с метриками для модели логистической регрессии.

## Кластеризация клиентов

### Стандартизация данных

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler() # создаём объект класса scaler (нормализатор)
X_sc = sc.fit_transform(X) # обучаем нормализатор и преобразуем набор данных
print(X_sc[:2]) 
```

### Матрица расстояний на стандартизованной матрице признаков, дендограмма

**Передаем получившуюся стандартизированную таблицу в качестве параметра функции linkage()**

```python
from scipy.cluster.hierarchy import dendrogram, linkage 
linked = linkage(X_sc, method = 'ward') 
```

**В переменной linked сохранена таблица «связок» между объектами. Её можно визуализировать как дендрограмму:**

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.show() 
```

На основании полученной дендограммы можно предположить, что можно выделить 5 кластеров.

### Обучение модели кластеризации на основании алгоритма K-Means и прогнозирование кластеров клиентов

```python
from sklearn.cluster import KMeans

# задаём модель k_means с числом кластеров 5
km = KMeans(n_clusters = 5)

# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 4)
labels = km.fit_predict(X_sc)

# сохраняем метки кластера в поле нашего датасета
df['cluster_km'] = labels
```

### Распределения признаков для кластеров

```python
for column in df.columns:
    plt.figure(figsize=(12, 5)) 
    sns.distplot(df[df['cluster_km']==0][column], bins=10, kde=False)
    sns.distplot(df[df['cluster_km']==1][column], bins=10, kde=False)
    sns.distplot(df[df['cluster_km']==2][column], bins=10, kde=False)
    sns.distplot(df[df['cluster_km']==3][column], bins=10, kde=False)
    sns.distplot(df[df['cluster_km']==4][column], bins=10, kde=False)
    plt.title(column)
    plt.show()
    
warnings.filterwarnings('ignore')
```

Рассмотрим наши признаки в разрезе кластеров:

* **near_location** - больше всех клиентов, которые близко живут к фитнес-центру в кластере 2;
* **partner** - больше всего клиентов из компаний партнеров в кластере 0;
* **promo_friends** - больше всего клиентов , пришедних по акции "Приведи подругу" в кластере 0;
* **phone** - не оставляет свой контактный номер телефона клиенты из кластера 2;
