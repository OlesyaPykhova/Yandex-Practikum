## Предобработка данных

### Проверка данных на пропуски, дубликаты, тип данных

```python
df.info()
```

В данных представлено 10000 строк с основной информацией по клиентам банка. Тип данных соответствует смысловой нагрузке каждого столбца - есть как количественные, так и категориальные признаки. Также видим, что **в столбце баланса на счете есть пропуски**, посмотрим, насколько их много в данных и после решим, что с ними делать.

```python
print('Количество дубликатов в данных: {}'.format(df.duplicated().sum()))
print('Доля пропущенных значений в данных: {}%'.format((df['balance'].isna().mean() * 100).round(1)))
```

**Проверим, есть ли клиенты с 0 болансом**

```python
df.query('balance == 0')
```

Пользователей с 0 балансом нет, можем предположить, что пропуски - это и есть 0 на счете. Так как пропущенных значений - треть данных, удалить мы их не можем без потери в результате анализа, поэтому заменим пропуски 0 и посмотрим на признаки для таких клиентов.

```python
df = df.fillna(0)   
df.query('balance == 0').head()
```

Среди клиентов, у которых 0 баланс на счете есть те, у кого оформлена кредитная карта в банке - можем предположить, что такие клиенты оформили только кредиторский счет, поэтому их баланс нулевой. Есть также и те, у кого не офорлмена кредитная карта, но видно, что они пользуются как минимум 1 продуктом банка - видимо, у таких клиентов также нет дебиторского счета. Оставляем заполнение пропусков 0 для дальнейшего анализа.

### Перевод категориальных признаков в бинарный код

```python
def gender_bul(row):
    if row['gender'] == 'Ж':
        return '0'
    else:
        return '1'

df['gender_bul'] = df.apply(gender_bul, axis=1)
df['gender_bul'] = df['gender_bul'].astype('int')
```

```python
df['city'].value_counts()
```

Категоризацию признака **'city'** проведем с помощью словаря, где 1 - Ярославль, 2 - Ростов Великий, 3 - Рыбинск.

```python
dct = {'Ярославль':1, 'Ростов Великий':2, 'Рыбинск':3}
df['city_numb'] = df['city'].map(dct)
df['city_numb'] = df['city_numb'].astype('int')
```

    - клиенты, которые пользуются 1 продуктом банка

```python
def segment(row):
    if row['products'] == 1:
        return '0'
    else:
        return '1'

data['segment'] = data.apply(segment, axis=1)
data['segment'] = data['segment'].astype('int')
data.head()
```

## Проверка статистических гипотез


### Проверка основной гипотезы

**Гипотеза H0** - доход пользователей, которые пользуются 1 продуктом банка НЕ отличается (то есть равен) доходу пользователей, которые пользуются 2 продуктами банка

**Гипотеза H1** - доход пользователей, которые пользуются 1 продуктом банка отличается от доходов пользователей, которые пользуются 2 продуктами банка

```python
one_product = data[data['products'] == 1]
two_product = data[data['products'] == 2]
```

```python
def sampling(data, column):
    plt.figure(figsize=(12, 5)) 
    print(data[column].describe())
    
    sns.distplot(data[column])
    plt.show()
    
sampling(one_product, 'estimated_salary')
sampling(two_product, 'estimated_salary')
```

**Проверка на выбросы**

```python
one_product.boxplot('estimated_salary')
plt.xlabel('Доход пользователей, которые пользуются 1 продуктом')
plt.show()

two_product.boxplot('estimated_salary')
plt.xlabel('Доход пользователей, которые пользуются 2 продуктами')
plt.show()
```

**Проверим дисперсии выборок**

```python
print(np.var(one_product['estimated_salary']))
print(np.var(two_product['estimated_salary']))
```

**Распределение признаков в каждой выборке соответствует нормальному распределению, стандартное отклонение признака в выборках также примерно равны. Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам,применим метод scipy.stats.ttest_ind()**

```python
one_product_sal = one_product['estimated_salary']
two_product_sal = two_product['estimated_salary']

from scipy import stats as st

results = st.ttest_ind(one_product_sal, two_product_sal, equal_var = True)
print('p-значение: ', results.pvalue)
print('пороговое значение alpha = 0.05')
```

***Гипотезу H0 отвергнуть не можем***, принимаем, что доход пользователей, которые пользуются 1 продуктом банка равен доходу пользователей, которые пользуются 2 продуктами банка

### Проверка дополнительной гипотезы 

**Гипотеза H0** - доход пользователей из 1 сегмента НЕ отличается (то есть равен) доходу пользователей из 2 сегмента

**Гипотеза H1** - доход пользователейиз 1 сегмента отличается от доходов пользователей из 2 сегмента

**Проверим дисперсии выборок**

## Модель прогнозирования оттока клиентов

### Обучение модели на train-выборке

**Разбиваем данные на обучающую и валидационную выборки**

```python
# Разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = data.drop('churn', axis = 1)
y = data['churn']

# Разбиваем данные на обучающую и валидационную выборки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

**Обучаем модель на train-выборке логистической регрессией**

```python
from sklearn.linear_model import LogisticRegression

# Задаем алгоритм для нашей модели
lr_model = LogisticRegression(random_state=0)

# Обучаем модель
lr_model.fit(X_train, y_train)

# Делаем прогноз для объектов из валидационного набора данных
lr_predictions = lr_model.predict(X_test)
lr_probabilities = lr_model.predict_proba(X_test)[:,1]

warnings.filterwarnings('ignore')
```

**Обучаем модель на train-выборке случайным лесом**

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Задаем алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)

# Обучаем модель случайного леса
rf_model.fit(X_train, y_train)

# Делаем прогнозы
rf_predictions = rf_model.predict(X_test)
rf_probabilities = rf_model.predict_proba(X_test)[:,1]
```

### Оценка и сравнение метрик на валидационной выборке

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score

# Функция,, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, y_proba, title = 'Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    
print_all_metrics(y_test, lr_predictions, lr_probabilities, title = 'Метрики для модели логистической регрессии:')
print_all_metrics(y_test, rf_predictions, rf_probabilities, title = 'Метрики для модели случайного леса:')
```

Метрики для модели случайного леса показали себя лучше в сравнении с метриками для модели логистической регрессии.
